```python
# Create a set
my_set = {1, 2, 3}

# Add a single element
my_set.add(4)  # Adds 4 to the set

# Add multiple elements (from list, tuple, or another set)
my_set.update([5, 6], {7, 8})

# Attempting to add a duplicate has no effect
my_set.add(3)

print(my_set)
```
